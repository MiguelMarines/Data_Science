# **Cliente de Python Firestore**

En esta fase se conectará a una base de datos no relacional de Firestore desde Python. Para ello se utilizarán los módulos `credentials` y `firestore` de la biblioteca `firebase_admin`.

```python
import firebase_admin
from firebase_admin import credentials
from firebase_admin import firestore
```

El archivo `veterinary.json` almacena la clave privada para autenticar una cuenta y autorizar el acceso a los servicios de Firebase. A través de la función `Certificate()`, se regresa una credencial inicializada, que se puede utilizar para crear una nueva instancia de la aplicación. Después de eso, la conexión a Firestore utilizará las reglas de seguridad establecidas para la base de datos y el usuario autenticado.

```python
cred = credentials.Certificate('veterinary.json')
firebase_admin.initialize_app(cred)
db = firestore.client()
```

```python
import pandas as pd
```

Lectura de la colección `PET_OWNER` y se muestra su contenido en un dataframe. Se incluye el id en el resultado.

```python
# Referencia a la colección PET_OWNER.
pet_owner_ref = db.collection("PET_OWNER")

# Se obtienen todos los documentos de la colección.
pet_owner_docs = pet_owner_ref.stream()

# Creación de una lista de diccionarios con los datos.
data = []
for doc in pet_owner_docs:
    doc_data = doc.to_dict()
    doc_data['ID'] = doc.id
    data.append(doc_data)

# Creación de un DataFrame a partir de los datos
df = pd.DataFrame(data)

# Orden de las columnas.
order = ['ID', 'ownerFirstName', 'ownerLastName', 'email', 'phone']
df = df[order]

# Se muestra el resultado.
df
```
