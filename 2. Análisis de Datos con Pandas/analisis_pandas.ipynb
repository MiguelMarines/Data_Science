10.	Se ejecuta el siguiente código y se comenta qué hace cada línea:

```python
# Creación de un nuevo DataFrame denominado "datetime_df", con la columna "ReadingDateTime" del DataFrame "air_df".
# La división de las columnas se determina por el espacio utilizado como separador entre la fecha y la hora.
datetime_df = air_df.ReadingDateTime.str.split(' ', expand = True)

# Las columnas del DataFrame "datetime_df" se nombran "Date" y "Time" correspondientemente.
datetime_df.columns = ['Date', 'Time']
# Nuevo DataFrame.
datetime_df

# Creación de un nuevo DataFrame denominado "date_df", con la columna "Date" del DataFrame "datetime_df".
# La división de las columnas se determina por la diagonal "/" utilizada como separador en la fecha.
date_df = datetime_df.Date.str.split('/', expand = True)
# Las columnas del DataFrame "date" se nombran "Day", "Month" y "Year" correspondientemente.
date_df.columns = ['Day', 'Month', 'Year']
# Nuevo DataFrame.
date_df

# Combina los dataframes 'date_df' y la columna del DataFrame 'datetime_df.Time' con el DataFrame original 'air_df'.
# Se eliminan las columnas 'ReadingDateTime' y 'Year' del DataFrame 'air_df'.
air_df = air_df.join(date_df).join(datetime_df.Time).drop(columns = ['ReadingDateTime', 'Year'])

# Establecimiento de un índice múltiple en base a las columnas de 'Month', 'Day', 'Time' y 'Species'.
air_df = air_df.set_index(['Month', 'Day', 'Time', 'Species'])
# Nuevo DataFrame.
air_df
```

12.	Explicación de las diferencias y similitudes entre el par `melt()/pivot()` y el par `stack()/unstack()`. Si se tuvieras que elegir una contraparte para `melt()` entre `stack()/unstack()`, ¿cuál se elegiría? ¿Por qué?

Diferencias y Similitudes entre melt() y pivot():
*   melt()

    * Transforma un DataFrame ancho (wide) en uno largo (long). Toma las columnas y las combina en dos columnas: una columna para las variables y otra para sus valores.
    * Es útil para desnormalizar datos, especialmente cuando se tienen múltiples columnas que representan diferentes categorías o medidas.
    * Requiere especificar una o más columnas que se mantendrán como identificadores (id_vars) y las columnas que se "desmelten" en una sola columna (value_vars).

*   pivot()

    * Conviertte un DataFrame largo (long) en uno ancho (wide), toma dos columnas (una para las variables y otra para sus valores) y las pivotea para formar un DataFrame con columnas separadas para cada variable.
    * Es útil para volver a estructurar datos en una forma más tabular y fácil de analizar.
    * Requiere especificar columnas para las filas (index), columnas para las columnas (columns), y valores para llenar las celdas.


Diferencias y Similitudes entre stack() y unstack():

*   stack()
    * Convierte columnas en niveles del índice. Toma las columnas del DataFrame y las apila en un índice multi-nivel.
    * Es útil para convertir columnas en índices, especialmente cuando se tienen varias columnas que representan diferentes categorías o niveles de una variable.
    * Es útil para reorganizar datos cuando se tienen índices multi-nivel.
*   unstack()
    * Desapila un índice multi-nivel y convierte uno de los niveles en columnas.
    * Es útil para revertir el efecto de stack() o para trabajar con datos de una forma más tabular.

Contraparte para melt() entre stack() y unstack():

*   ¿Cuál se elegiría? y ¿Por qué?: Elegiría unstack(), ya que melt() se utiliza para convertir un DataFrame ancho en uno largo, mientras que unstack() se utiliza para revertir el efecto de stack() y convertir un índice multi-nivel en columnas y volver a estructurar los datos en una forma tabular que sea más conveniente para el análisis y la visualización.
